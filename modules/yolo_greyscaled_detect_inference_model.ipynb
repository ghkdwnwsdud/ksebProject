```python
import cv2
import matplotlib.pyplot as plt
from ultralytics import YOLO
from google.colab.patches import cv2_imshow
import torch
import torchvision.transforms as transforms
import numpy as np
import os

# 객체 탐지 함수
def detect_objects_yolo(model, frame, conf_threshold=0.3):
    # 이미지를 Torch 텐서로 변환
    # 모델에 이미지 입력 후 추론
    results = model(frame)  # 추론

    # 추론 결과에서 객체 정보 추출
    detected_objects = []
    for box in results[0].boxes:
        class_id = int(box.cls)
        confidence = box.conf.item()  # 텐서를 float로 변환
        class_name = model.names[class_id]


        if confidence > conf_threshold:
            # 바운딩 박스 좌표 계산
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            obj_img = frame[y1:y2, x1:x2]  # 객체 부분 이미지 추출
            detected_objects.append((class_id, obj_img))
            cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)
            label = f'{class_name}: {confidence:.2f}'
            cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

    return detected_objects

# 1. 훈련된 모델 로드
yolo_model = YOLO('/content/best (1).pt') #본인의 훈련된 파일 경로


# 입력 동영상 파일 경로
video_path = '/content/videoplayback.mp4'

#폴더 경로
class_a_folder = '/content/drive/MyDrive/Colab Notebooks/부트캠프/투수_타자/hitter'
class_b_folder = '/content/drive/MyDrive/Colab Notebooks/부트캠프/투수_타자/pitcher'

# 동영상 파일 열기
cap = cv2.VideoCapture(video_path)

frame_interval = 5  # 매 5프레임마다 한 번씩 처리

# 초기 프레임 카운터
frame_count = 0

# 프레임 순회
while True:
    ret, frame = cap.read()
    if not ret:
        break

    # 객체 탐지 수행
    detected_objects = detect_objects_yolo(yolo_model, frame)

    # 각 객체 이미지 저장
    for class_id, obj_img in detected_objects:
        if obj_img is not None:
            if class_id == 0:
                cv2.imwrite(os.path.join(class_a_folder, f'object_{len(os.listdir(class_a_folder))}.jpg'), obj_img)
            elif class_id == 1:
                cv2.imwrite(os.path.join(class_b_folder, f'object_{len(os.listdir(class_b_folder))}.jpg'), obj_img)

    # 화면에 출력
    #cv2_imshow(frame)
    #if cv2.waitKey(1) & 0xFF == ord('q'):
    #    break

# 종료
cap.release()
cv2.destroyAllWindows()
```
