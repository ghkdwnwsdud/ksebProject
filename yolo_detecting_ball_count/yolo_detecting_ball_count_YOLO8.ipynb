```python
import os
import cv2
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# ÎîîÎ†âÌÜ†Î¶¨ ÏÑ§Ï†ï
image_dir = '/content/drive/MyDrive/KSEB2/new/output/regions'
csv_path = '/content/drive/MyDrive/KSEB2/label_final.csv'

# CSV ÌååÏùº Î°úÎìú
df = pd.read_csv(csv_path)

# Îç∞Ïù¥ÌÑ∞ Ï§ÄÎπÑ
images = []
labels = []

# ÏµúÎåÄ ÌÅ¨Í∏∞ Í≥ÑÏÇ∞
max_height = 0
max_width = 0

for index, row in df.iterrows():
    image_path = os.path.join(image_dir, row['filename'])
    image = cv2.imread(image_path)
    if image is not None:
        h, w, _ = image.shape
        if h > max_height:
            max_height = h
        if w > max_width:
            max_width = w

# Ïù¥ÎØ∏ÏßÄ Ìå®Îî© Ìï®Ïàò
def pad_image(image, target_height, target_width):
    height, width, _ = image.shape
    padded_image = np.zeros((target_height, target_width, 3), dtype=np.uint8)
    padded_image[:height, :width, :] = image
    return padded_image

for index, row in df.iterrows():
    image_path = os.path.join(image_dir, row['filename'])
    image = cv2.imread(image_path)
    if image is not None:
        padded_image = pad_image(image, max_height, max_width)
        images.append(padded_image)
        labels.append([row['ball'], row['strike'], row['out']])

images = np.array(images)
labels = np.array(labels)

# Îç∞Ïù¥ÌÑ∞ÏÖã Î∂ÑÌï†
X_train, X_test, y_train, y_test = train_test_split(images, labels, test_size=0.2, random_state=42)

# Î™®Îç∏ Íµ¨ÏÑ±
model = Sequential([
    Conv2D(32, (2, 2), activation='relu', input_shape=(max_height, max_width, 3)),
    MaxPooling2D((1, 1)),
    Conv2D(64, (2, 2), activation='relu'),
    MaxPooling2D((1, 1)),
    Conv2D(128, (2, 2), activation='relu'),
    MaxPooling2D((1, 1)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(3)  # 3Í∞úÏùò Ï∂úÎ†•ÏùÑ ÏúÑÌïú Dense Î†àÏù¥Ïñ¥ (ball, strike, out)
])

# Î™®Îç∏ Ïª¥ÌååÏùº
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])

# Îç∞Ïù¥ÌÑ∞ Ï¶ùÍ∞ï Î∞è Î∞∞Ïπò ÏÉùÏÑ±Í∏∞
train_datagen = tf.keras.preprocessing.image.ImageDataGenerator()
train_generator = train_datagen.flow(X_train, y_train, batch_size=32)

# Î™®Îç∏ ÌïôÏäµ
model.fit(train_generator, epochs=20, validation_data=(X_test, y_test))

# Î™®Îç∏ ÌèâÍ∞Ä
loss, mae = model.evaluate(X_test, y_test)
print(f'Test MAE: {mae}')



```

```python
import os
import pandas as pd
import cv2
import numpy as np
from sklearn.model_selection import train_test_split

# Ïù¥ÎØ∏ÏßÄ Í≤ΩÎ°úÏôÄ CSV ÌååÏùº Í≤ΩÎ°ú ÏÑ§Ï†ï
image_dir = '/content/drive/MyDrive/KSEB2/new/output/regions'
csv_path = '/content/drive/MyDrive/KSEB2/label_final.csv'

# CSV ÌååÏùº Î°úÎìú
df = pd.read_csv(csv_path)

# Ïù¥ÎØ∏ÏßÄ ÌÅ¨Í∏∞ ÏÑ§Ï†ï
IMG_WIDTH = 64
IMG_HEIGHT = 64

# Ïù¥ÎØ∏ÏßÄ Îç∞Ïù¥ÌÑ∞ÏôÄ Î†àÏù¥Î∏î Ï§ÄÎπÑ
images = []
labels = []

for idx, row in df.iterrows():
    image_path = os.path.join(image_dir, row['filename'])
    image = cv2.imread(image_path)
    if image is not None:
        image = cv2.resize(image, (IMG_WIDTH, IMG_HEIGHT))  # Ïù¥ÎØ∏ÏßÄ Î¶¨ÏÇ¨Ïù¥Ï¶à
        image = image / 255.0  # Ïù¥ÎØ∏ÏßÄ Ï†ïÍ∑úÌôî
        images.append(image)
        labels.append([row['ball'], row['strike'], row['out']])

# numpy Î∞∞Ïó¥Î°ú Î≥ÄÌôò
images = np.array(images)
labels = np.array(labels)

# ÌïôÏäµ Îç∞Ïù¥ÌÑ∞ÏôÄ Í≤ÄÏ¶ù Îç∞Ïù¥ÌÑ∞Î°ú Î∂ÑÌï†
X_train, X_val, y_train, y_val = train_test_split(images, labels, test_size=0.2, random_state=42)
```

```python
import cv2
import os
import pandas as pd
from IPython.display import display, clear_output
from google.colab.patches import cv2_imshow

# ÎîîÎ†âÌÜ†Î¶¨ ÏÑ§Ï†ï
image_dir = '/content/drive/MyDrive/KSEB2/2/output/regions'
train_image_dir = '/content/drive/MyDrive/KSEB2/2/dataset/train/images'
train_label_dir = '/content/drive/MyDrive/KSEB2/2/dataset/train/labels'
csv_path = '/content/drive/MyDrive/KSEB2/2/labels.csv'

# CSV ÌååÏùº Î°úÎìú Î∞è Îç∞Ïù¥ÌÑ∞ ÌÉÄÏûÖ ÏßÄÏ†ï
df = pd.read_csv(csv_path, dtype={'filename': str, 'ball': float, 'strike': float, 'out': float})

# NaN Í∞íÏù¥ ÏûàÎäî Ìñâ ÌïÑÌÑ∞ÎßÅ
nan_rows = df[df[['ball', 'strike', 'out']].isna().any(axis=1)]
initial_remaining_files_count = len(nan_rows)

for idx in nan_rows.index:
    row = df.loc[idx]
    image_file = row['filename']
    ball = row['ball']
    strike = row['strike']
    out = row['out']

    image_path = os.path.join(image_dir, image_file)
    image = cv2.imread(image_path)

    if image is None:
        continue

    # Ïù¥ÎØ∏ÏßÄ ÌÅ¨Í∏∞ Ï°∞Ï†ï (3Î∞∞Î°ú ÌôïÎåÄ)
    height, width = image.shape[:2]
    scale_factor = 3  # Ïù¥ÎØ∏ÏßÄ ÌÅ¨Í∏∞Î•º 3Î∞∞Î°ú ÌôïÎåÄ
    resized_image = cv2.resize(image, (width * scale_factor, height * scale_factor))

    # Ïù¥ÎØ∏ÏßÄÏôÄ Ï†ïÎ≥¥ ÌôîÎ©¥Ïóê ÌëúÏãú
    clear_output(wait=True)
    print(f"ÌòÑÏû¨ Îç∞Ïù¥ÌÑ∞ ÌîÑÎ†àÏûÑ Ìñâ Ïàò: {len(df)}")
    print(f"ÌôïÏù∏Ìï¥ÏïºÌï† ÎÇ®ÏùÄ ÌååÏùº Ïàò: {initial_remaining_files_count}")
    print(f"Displaying {image_file}")
    cv2_imshow(resized_image)
    print(f"balls: {ball}, strikes: {strike}, outs: {out}")
    print("Enter to keep, 'x' to delete, 'e' to edit: ")
    action = input("").strip().lower()

    if action == 'x':
        # Ïù¥ÎØ∏ÏßÄ ÌååÏùº ÏÇ≠Ï†ú
        os.remove(image_path)

        # train/images Ìè¥ÎçîÏùò ÌååÏùº ÏÇ≠Ï†ú
        train_image_path = os.path.join(train_image_dir, image_file)
        if os.path.exists(train_image_path):
            os.remove(train_image_path)

        # train/labels Ìè¥ÎçîÏùò ÌååÏùº ÏÇ≠Ï†ú
        label_file = image_file.replace('.jpg', '.txt').replace('.JPG', '.txt')
        train_label_path = os.path.join(train_label_dir, label_file)
        if os.path.exists(train_label_path):
            os.remove(train_label_path)

        # Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏóêÏÑú Ìï¥Îãπ Ìñâ ÏÇ≠Ï†ú
        df = df.drop(idx)
        # Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ Ïù∏Îç±Ïä§ Ïû¨ÏÑ§Ï†ï
        df = df.reset_index(drop=True)
    elif action == 'e':
        # ÏàòÏ†ïÌï† Í∞í ÏûÖÎ†• Î∞õÍ∏∞
        ball = float(input("new balls: ").strip())
        strike = int(input("new strikes: ").strip())
        out = float(input("new outs: ").strip())
        # Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑ ÏóÖÎç∞Ïù¥Ìä∏
        df.at[idx, 'ball'] = ball
        df.at[idx, 'strike'] = strike
        df.at[idx, 'out'] = out

    # ÎÇ®ÏùÄ ÌååÏùº Ïàò ÏóÖÎç∞Ïù¥Ìä∏
    initial_remaining_files_count -= 1

    # CSV ÌååÏùºÏóê Ï†ÄÏû•
    df.to_csv(csv_path, index=False)

cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import itertools
import os

# ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞ Í∑∏Î¶¨Îìú Ï†ïÏùò
lr_values = [0.0001, 0.0005, 0.001]
momentum_values = [0.9, 0.937, 0.99]
weight_decay_values = [0.0001, 0.0005, 0.001]
batch_sizes = [256, 128, 64, 32]

# Í∞ÄÎä•Ìïú ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞ Ï°∞Ìï© ÏÉùÏÑ±
param_grid = list(itertools.product(lr_values, momentum_values, weight_decay_values, batch_sizes))

# Îç∞Ïù¥ÌÑ∞ÏÖã Î∞è Í∏∞ÌÉÄ ÏÑ§Ï†ï
yaml_file_path = '/path/to/your/data.yaml'
output_dir = '/content/drive/MyDrive/KSEB2/final3/output'
epochs = 50
imgsz = 640
patience = 3  # Ï°∞Í∏∞ Ï¢ÖÎ£åÎ•º ÏúÑÌïú patience ÏÑ§Ï†ï

# ÏµúÏ†ÅÏùò ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞ Ï°∞Ìï© Ï∞æÍ∏∞
best_performance = float('inf')
best_params = None

for lr, momentum, weight_decay, batch_size in param_grid:
    # Î™®Îç∏ Î°úÎìú
    model = YOLO('/content/drive/MyDrive/KSEB2/final3/best_new.pt')

    # ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞ ÏÑ§Ï†ï
    hyp = {
        'lr0': lr,
        'momentum': momentum,
        'weight_decay': weight_decay
    }

    # ÌïôÏäµ
    results = model.train(data=yaml_file_path, epochs=epochs, imgsz=imgsz, batch=batch_size, name=f'yolo_v8_training_lr{lr}_mom{momentum}_wd{weight_decay}_bs{batch_size}', save_dir=output_dir, patience=patience, **hyp)

    # Í≤ÄÏ¶ù ÏÑ±Îä• ÌèâÍ∞Ä
    val_results = model.val()
    performance = val_results.results_dict['metrics']['mAP_50_95']  # Ïòà: mAP ÎòêÎäî Îã§Î•∏ ÏÑ±Îä• ÏßÄÌëú

    # ÏµúÏ†ÅÏùò ÌïòÏù¥ÌçºÌååÎùºÎØ∏ÌÑ∞ Ï°∞Ìï© ÏóÖÎç∞Ïù¥Ìä∏
    if performance < best_performance:
        best_performance = performance
        best_params = (lr, momentum, weight_decay, batch_size)

print(f"Best Performance: {best_performance}")
print(f"Best Hyperparameters: lr={best_params[0]}, momentum={best_params[1]}, weight_decay={best_params[2]}, batch_size={best_params[3]}")
```

```python
import cv2
import os
from ultralytics import YOLO

# Î™®Îç∏ Î°úÎìú
model = YOLO('/content/drive/MyDrive/KSEB2/final3/yolo_v8_training_lr0.0005_mom0.9_wd0.0001_bs32_last.pt')  # ÌïôÏäµÎêú Î™®Îç∏ Í≤ΩÎ°ú

# ÏûÖÎ†• ÎπÑÎîîÏò§ Í≤ΩÎ°úÏôÄ Ï∂úÎ†• ÎπÑÎîîÏò§ Í≤ΩÎ°ú ÏÑ§Ï†ï
input_video_path = '/content/drive/MyDrive/KSEB2/final3/baseball_play2.mp4'
output_video_path = '/content/drive/MyDrive/KSEB2/final3/baseball_play2_output_LR0.005_bs32.mp4'

# ÎπÑÎîîÏò§ Ï∫°Ï≤ò Í∞ùÏ≤¥ ÏÉùÏÑ±
cap = cv2.VideoCapture(input_video_path)
width = 640  # ÏÉàÎ°úÏö¥ ÌîÑÎ†àÏûÑ ÎÑàÎπÑ
height = 640  # ÏÉàÎ°úÏö¥ ÌîÑÎ†àÏûÑ ÎÜíÏù¥
fps = cap.get(cv2.CAP_PROP_FPS)
fourcc = cv2.VideoWriter_fourcc(*'mp4v')

# ÎπÑÎîîÏò§ ÎùºÏù¥ÌÑ∞ Í∞ùÏ≤¥ ÏÉùÏÑ±
out = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # ÌîÑÎ†àÏûÑ ÌÅ¨Í∏∞ Ï°∞Ï†ï
    frame = cv2.resize(frame, (width, height))

    # YOLOv8 Î™®Îç∏Î°ú ÌîÑÎ†àÏûÑ Í≤ÄÏ∂ú
    results = model(frame)

    # Í≤ÄÏ∂ú Í≤∞Í≥ºÎ•º ÌîÑÎ†àÏûÑÏóê Ï†ÅÏö©
    for result in results:
        boxes = result.boxes.xyxy.cpu().numpy()  # bounding box Ï¢åÌëú
        confidences = result.boxes.conf.cpu().numpy()  # Ïã†Î¢∞ÎèÑ
        class_ids = result.boxes.cls.cpu().numpy()  # ÌÅ¥ÎûòÏä§ ID

        for box, confidence, class_id in zip(boxes, confidences, class_ids):
            x1, y1, x2, y2 = map(int, box)
            label = f"{model.names[int(class_id)]} {confidence:.2f}"
            color = (0, 255, 0)  # bounding box ÏÉâÏÉÅ (Ï¥àÎ°ùÏÉâ)

            # bounding boxÏôÄ ÎùºÎ≤®ÏùÑ ÌîÑÎ†àÏûÑÏóê Í∑∏Î¶¨Í∏∞
            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
            cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

    # Í≤∞Í≥º ÌîÑÎ†àÏûÑÏùÑ ÎπÑÎîîÏò§ ÎùºÏù¥ÌÑ∞Ïóê ÏûëÏÑ±
    out.write(frame)

# Î¶¨ÏÜåÏä§ Ìï¥Ï†ú
cap.release()
out.release()
cv2.destroyAllWindows()

print(f"Output video saved to {output_video_path}")
```
