```python
import os
import cv2
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# 디렉토리 설정
image_dir = '/content/drive/MyDrive/KSEB2/new/output/regions'
csv_path = '/content/drive/MyDrive/KSEB2/label_final.csv'

# CSV 파일 로드
df = pd.read_csv(csv_path)

# 데이터 준비
images = []
labels = []

# 최대 크기 계산
max_height = 0
max_width = 0

for index, row in df.iterrows():
    image_path = os.path.join(image_dir, row['filename'])
    image = cv2.imread(image_path)
    if image is not None:
        h, w, _ = image.shape
        if h > max_height:
            max_height = h
        if w > max_width:
            max_width = w

# 이미지 패딩 함수
def pad_image(image, target_height, target_width):
    height, width, _ = image.shape
    padded_image = np.zeros((target_height, target_width, 3), dtype=np.uint8)
    padded_image[:height, :width, :] = image
    return padded_image

for index, row in df.iterrows():
    image_path = os.path.join(image_dir, row['filename'])
    image = cv2.imread(image_path)
    if image is not None:
        padded_image = pad_image(image, max_height, max_width)
        images.append(padded_image)
        labels.append([row['ball'], row['strike'], row['out']])

images = np.array(images)
labels = np.array(labels)

# 데이터셋 분할
X_train, X_test, y_train, y_test = train_test_split(images, labels, test_size=0.2, random_state=42)

# 모델 구성
model = Sequential([
    Conv2D(32, (2, 2), activation='relu', input_shape=(max_height, max_width, 3)),
    MaxPooling2D((1, 1)),
    Conv2D(64, (2, 2), activation='relu'),
    MaxPooling2D((1, 1)),
    Conv2D(128, (2, 2), activation='relu'),
    MaxPooling2D((1, 1)),
    Flatten(),
    Dense(128, activation='relu'),
    Dense(3)  # 3개의 출력을 위한 Dense 레이어 (ball, strike, out)
])

# 모델 컴파일
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])

# 데이터 증강 및 배치 생성기
train_datagen = tf.keras.preprocessing.image.ImageDataGenerator()
train_generator = train_datagen.flow(X_train, y_train, batch_size=32)

# 모델 학습
model.fit(train_generator, epochs=20, validation_data=(X_test, y_test))

# 모델 평가
loss, mae = model.evaluate(X_test, y_test)
print(f'Test MAE: {mae}')



```

```python
from ultralytics import YOLO
import os
import cv2
import numpy as np
import pandas as pd

# 모델 경로 설정
yolo_model_path = '/content/drive/MyDrive/KSEB2/bes.pt'
region_model_path = '/content/drive/MyDrive/KSEB2/saved_model/my_model.h5'

# YOLOv8 모델 로드
if os.path.exists(yolo_model_path):
    yolo_model = YOLO(yolo_model_path)
else:
    print(f"YOLO model file not found at {yolo_model_path}")

# 지역 분류기 모델 로드
if os.path.exists(region_model_path):
    region_model = load_model(region_model_path)
else:
    print(f"Region classifier model file not found at {region_model_path}")

# 이미지 크기 조정 함수
def resize_image(image, target_height, target_width):
    return cv2.resize(image, (target_width, target_height))

# 새로운 이미지에 대해 예측
def predict_ball_strike_out(image_path):
    image = cv2.imread(image_path)
    input_height = 17
    input_width = 78
    image = resize_image(image, input_height, input_width)
    image = np.expand_dims(image, axis=0)  # 배치 차원 추가
    prediction = region_model.predict(image)
    ball, strike, out = np.round(prediction[0]).astype(int)  # 소수점 값을 정수로 반올림하여 변환
    ball = np.clip(ball, 0, 3)  # ball 값을 0에서 3 사이로 제한
    strike = np.clip(strike, 0, 2)  # strike 값을 0에서 2 사이로 제한
    out = np.clip(out, 0, 2)  # out 값을 0에서 2 사이로 제한
    return ball, strike, out

# /content/drive/MyDrive/KSEB2/3/output/regions 폴더의 모든 이미지에 대한 예측
prediction_results = []
regions_dir = '/content/drive/MyDrive/KSEB2/3/output/regions'
for filename in os.listdir(regions_dir):
    if filename.endswith(('.jpg', '.png')):
        file_path = os.path.join(regions_dir, filename)
        ball, strike, out = predict_ball_strike_out(file_path)
        prediction_results.append({
            'filename': filename,
            'ball': ball,
            'strike': strike,
            'out': out
        })

# 예측 결과 출력
for result in prediction_results:
    print(f"File: {result['filename']}, Ball: {result['ball']}, Strike: {result['strike']}, Out: {result['out']}")

# 실제 결과 로드
actual_csv_path = '/content/drive/MyDrive/KSEB2/3/labels.csv'
actual_df = pd.read_csv(actual_csv_path)

# 정확도 계산
correct_predictions = 0
total_predictions = 0

for result in prediction_results:
    filename = result['filename']
    actual_row = actual_df[actual_df['filename'] == filename]
    if not actual_row.empty:
        actual_ball = actual_row['ball'].values[0]
        actual_strike = actual_row['strike'].values[0]
        actual_out = actual_row['out'].values[0]
        if (result['ball'] == actual_ball) and (result['strike'] == actual_strike) and (result['out'] == actual_out):
            correct_predictions += 1
        total_predictions += 1

accuracy = (correct_predictions / total_predictions) * 100 if total_predictions > 0 else 0
print(f"Accuracy: {accuracy:.2f}%")
```

```python
from ultralytics import YOLO
import cv2
import numpy as np
from tensorflow.keras.models import load_model
from google.colab.patches import cv2_imshow

# YOLOv8 모델 로드
yolo_model = YOLO('/content/drive/MyDrive/KSEB2/bes.pt')

# 지역 분류기 모델 로드
region_model = load_model('/content/drive/MyDrive/KSEB2/saved_model/my_model.h5')

# 전체 이미지 예측
def detect_and_classify(image_path):
    # 전체 이미지 로드
    image = cv2.imread(image_path)
    if image is None:
        print(f"Error loading image: {image_path}")
        return

    # YOLO 모델로 예측
    results = yolo_model.predict(image_path)

    # 탐지된 모든 영역에 대해 예측
    for result in results:
        if len(result.boxes) == 0:
            print(f"No boxes detected in image: {image_path}")
            continue

        for box in result.boxes:
            x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())
            region = image[y1:y2, x1:x2]

            # 영역 이미지를 리사이즈 및 전처리
            region = cv2.resize(region, (78, 17))  # 모델이 기대하는 크기
            region = region / 255.0
            region = np.expand_dims(region, axis=0)

            # 지역 분류기 모델로 예측
            prediction = region_model.predict(region)
            ball, strike, out = map(int, np.round(prediction[0]))

            # 탐지된 영역을 원본 이미지에 그리기
            cv2.rectangle(image, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(image, f'B:{ball} S:{strike} O:{out}', (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (36, 255, 12), 2)

            # 결과 출력
            print(f'Ball: {ball}, Strike: {strike}, Out: {out}')

    # 이미지 화면에 표시
    cv2_imshow(image)

# 예시 이미지에서 예측
image_path = '/content/drive/MyDrive/KSEB2/3/dataset/train/images/-207-_png.rf.26597b885c4677b691456f0b0ebaf3da.jpg'
detect_and_classify(image_path)
```

```python
import os
import pandas as pd
import cv2
import numpy as np
from sklearn.model_selection import train_test_split

# 이미지 경로와 CSV 파일 경로 설정
image_dir = '/content/drive/MyDrive/KSEB2/new/output/regions'
csv_path = '/content/drive/MyDrive/KSEB2/label_final.csv'

# CSV 파일 로드
df = pd.read_csv(csv_path)

# 이미지 크기 설정
IMG_WIDTH = 64
IMG_HEIGHT = 64

# 이미지 데이터와 레이블 준비
images = []
labels = []

for idx, row in df.iterrows():
    image_path = os.path.join(image_dir, row['filename'])
    image = cv2.imread(image_path)
    if image is not None:
        image = cv2.resize(image, (IMG_WIDTH, IMG_HEIGHT))  # 이미지 리사이즈
        image = image / 255.0  # 이미지 정규화
        images.append(image)
        labels.append([row['ball'], row['strike'], row['out']])

# numpy 배열로 변환
images = np.array(images)
labels = np.array(labels)

# 학습 데이터와 검증 데이터로 분할
X_train, X_val, y_train, y_val = train_test_split(images, labels, test_size=0.2, random_state=42)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

# 모델 정의
model = Sequential([
    Conv2D(32, (3, 3), activation='relu', input_shape=(IMG_WIDTH, IMG_HEIGHT, 3)),
    MaxPooling2D((2, 2)),
    Conv2D(64, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Conv2D(128, (3, 3), activation='relu'),
    MaxPooling2D((2, 2)),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(3)  # ball, strike, out 예측을 위한 출력층
])

model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])

# 모델 학습
history = model.fit(X_train, y_train, epochs=50, validation_data=(X_val, y_val), batch_size=32)

# 모델 저장 (최신 Keras 형식으로 저장)
model.save('/content/drive/MyDrive/KSEB2/saved_model/my_model22.keras')
```

```python
import cv2
import os
import pandas as pd
from IPython.display import display, clear_output
from google.colab.patches import cv2_imshow

# 디렉토리 설정
image_dir = '/content/drive/MyDrive/KSEB2/2/output/regions'
train_image_dir = '/content/drive/MyDrive/KSEB2/2/dataset/train/images'
train_label_dir = '/content/drive/MyDrive/KSEB2/2/dataset/train/labels'
csv_path = '/content/drive/MyDrive/KSEB2/2/labels.csv'

# CSV 파일 로드 및 데이터 타입 지정
df = pd.read_csv(csv_path, dtype={'filename': str, 'ball': float, 'strike': float, 'out': float})

# NaN 값이 있는 행 필터링
nan_rows = df[df[['ball', 'strike', 'out']].isna().any(axis=1)]
initial_remaining_files_count = len(nan_rows)

for idx in nan_rows.index:
    row = df.loc[idx]
    image_file = row['filename']
    ball = row['ball']
    strike = row['strike']
    out = row['out']

    image_path = os.path.join(image_dir, image_file)
    image = cv2.imread(image_path)

    if image is None:
        continue

    # 이미지 크기 조정 (3배로 확대)
    height, width = image.shape[:2]
    scale_factor = 3  # 이미지 크기를 3배로 확대
    resized_image = cv2.resize(image, (width * scale_factor, height * scale_factor))

    # 이미지와 정보 화면에 표시
    clear_output(wait=True)
    print(f"현재 데이터 프레임 행 수: {len(df)}")
    print(f"확인해야할 남은 파일 수: {initial_remaining_files_count}")
    print(f"Displaying {image_file}")
    cv2_imshow(resized_image)
    print(f"balls: {ball}, strikes: {strike}, outs: {out}")
    print("Enter to keep, 'x' to delete, 'e' to edit: ")
    action = input("").strip().lower()

    if action == 'x':
        # 이미지 파일 삭제
        os.remove(image_path)

        # train/images 폴더의 파일 삭제
        train_image_path = os.path.join(train_image_dir, image_file)
        if os.path.exists(train_image_path):
            os.remove(train_image_path)

        # train/labels 폴더의 파일 삭제
        label_file = image_file.replace('.jpg', '.txt').replace('.JPG', '.txt')
        train_label_path = os.path.join(train_label_dir, label_file)
        if os.path.exists(train_label_path):
            os.remove(train_label_path)

        # 데이터프레임에서 해당 행 삭제
        df = df.drop(idx)
        # 데이터프레임 인덱스 재설정
        df = df.reset_index(drop=True)
    elif action == 'e':
        # 수정할 값 입력 받기
        ball = float(input("new balls: ").strip())
        strike = int(input("new strikes: ").strip())
        out = float(input("new outs: ").strip())
        # 데이터프레임 업데이트
        df.at[idx, 'ball'] = ball
        df.at[idx, 'strike'] = strike
        df.at[idx, 'out'] = out

    # 남은 파일 수 업데이트
    initial_remaining_files_count -= 1

    # CSV 파일에 저장
    df.to_csv(csv_path, index=False)

cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import itertools
import os

# 하이퍼파라미터 그리드 정의
lr_values = [0.0001, 0.0005, 0.001]
momentum_values = [0.9, 0.937, 0.99]
weight_decay_values = [0.0001, 0.0005, 0.001]
batch_sizes = [256, 128, 64, 32]

# 가능한 하이퍼파라미터 조합 생성
param_grid = list(itertools.product(lr_values, momentum_values, weight_decay_values, batch_sizes))

# 데이터셋 및 기타 설정
yaml_file_path = '/path/to/your/data.yaml'
output_dir = '/content/drive/MyDrive/KSEB2/final3/output'
epochs = 50
imgsz = 640
patience = 3  # 조기 종료를 위한 patience 설정

# 최적의 하이퍼파라미터 조합 찾기
best_performance = float('inf')
best_params = None

for lr, momentum, weight_decay, batch_size in param_grid:
    # 모델 로드
    model = YOLO('/content/drive/MyDrive/KSEB2/final3/best_new.pt')

    # 하이퍼파라미터 설정
    hyp = {
        'lr0': lr,
        'momentum': momentum,
        'weight_decay': weight_decay
    }

    # 학습
    results = model.train(data=yaml_file_path, epochs=epochs, imgsz=imgsz, batch=batch_size, name=f'yolo_v8_training_lr{lr}_mom{momentum}_wd{weight_decay}_bs{batch_size}', save_dir=output_dir, patience=patience, **hyp)

    # 검증 성능 평가
    val_results = model.val()
    performance = val_results.results_dict['metrics']['mAP_50_95']  # 예: mAP 또는 다른 성능 지표

    # 최적의 하이퍼파라미터 조합 업데이트
    if performance < best_performance:
        best_performance = performance
        best_params = (lr, momentum, weight_decay, batch_size)

print(f"Best Performance: {best_performance}")
print(f"Best Hyperparameters: lr={best_params[0]}, momentum={best_params[1]}, weight_decay={best_params[2]}, batch_size={best_params[3]}")
```

### 결과 확인

```python
import cv2
import os
from ultralytics import YOLO

# 모델 로드
model = YOLO('/content/drive/MyDrive/KSEB2/final3/yolo_v8_training_lr0.0005_mom0.9_wd0.0001_bs32_last.pt')  # 학습된 모델 경로

# 입력 비디오 경로와 출력 비디오 경로 설정
input_video_path = '/content/drive/MyDrive/KSEB2/final3/baseball_play2.mp4'
output_video_path = '/content/drive/MyDrive/KSEB2/final3/baseball_play2_output_LR0.005_bs32.mp4'

# 비디오 캡처 객체 생성
cap = cv2.VideoCapture(input_video_path)
width = 640  # 새로운 프레임 너비
height = 640  # 새로운 프레임 높이
fps = cap.get(cv2.CAP_PROP_FPS)
fourcc = cv2.VideoWriter_fourcc(*'mp4v')

# 비디오 라이터 객체 생성
out = cv2.VideoWriter(output_video_path, fourcc, fps, (width, height))

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # 프레임 크기 조정
    frame = cv2.resize(frame, (width, height))

    # YOLOv8 모델로 프레임 검출
    results = model(frame)

    # 검출 결과를 프레임에 적용
    for result in results:
        boxes = result.boxes.xyxy.cpu().numpy()  # bounding box 좌표
        confidences = result.boxes.conf.cpu().numpy()  # 신뢰도
        class_ids = result.boxes.cls.cpu().numpy()  # 클래스 ID

        for box, confidence, class_id in zip(boxes, confidences, class_ids):
            x1, y1, x2, y2 = map(int, box)
            label = f"{model.names[int(class_id)]} {confidence:.2f}"
            color = (0, 255, 0)  # bounding box 색상 (초록색)

            # bounding box와 라벨을 프레임에 그리기
            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
            cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

    # 결과 프레임을 비디오 라이터에 작성
    out.write(frame)

# 리소스 해제
cap.release()
out.release()
cv2.destroyAllWindows()

print(f"Output video saved to {output_video_path}")
```
